# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

```python
def is_even_or_odd(number):
    # This operation takes a fixed amount of time, regardless of the size of 'number'
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    # The actual value of 'n' doesn't affect the execution time of is_even_or_odd
    is_even_or_odd(n)
    time_values.append(1) # Representing constant time for each operation

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Even/Odd Check")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
def count_elements_greater_than_zero(numbers):
    count = 0
    for num in numbers:
        if num > 0:
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    # Create a list with 'n' elements for testing
    data = list(range(-n // 2, n // 2)) # Example data with positive and negative numbers
    count_elements_greater_than_zero(data)
    time_values.append(n) # Representing linear time for each operation

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity for Counting Positive Elements")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
import math

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    steps = 0
    while low <= high:
        steps += 1 # Increment step count for each iteration
        mid = (low + high) // 2
        if arr[mid] == target:
            return steps # Target found
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps # Target not found, return total steps

n_values = []
time_values = []

# We'll use powers of 2 for n to clearly show logarithmic behavior
# For larger n values, the log behavior becomes more evident
for i in range(1, 15):
    n = 2**i # Input size (e.g., 2, 4, 8, 16, ..., 16384)
    n_values.append(n)

    # Create a sorted array of size n
    data_array = list(range(n))

    # For binary search, the 'worst case' number of comparisons is log2(n)
    # We'll simulate this by searching for a value not present or at the ends
    # to approximate the log(n) behavior.
    # Alternatively, we can just use math.log2(n) as the time complexity
    time_values.append(math.log2(n))

plt.figure()
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Logarithmic Time (log n)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search Simulation)")
plt.grid(True)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
import math
import time
import random

def log_linear_sort(data):
    # Python's built-in sort is typically Timsort, which is O(n log n)
    # We'll just perform the sort operation and record the conceptual time
    data.sort()
    return

n_values = []
time_values = []

# We'll use a range of n values to demonstrate the growth
# Start with slightly larger values for n to make the log n factor more significant
for i in range(1, 13):
    n = 2**i * 100 # Input size (e.g., 200, 400, 800, ..., 204800)
    n_values.append(n)

    # Simulate O(n log n) time complexity
    # The actual execution time would be harder to measure consistently
    # so we'll plot the theoretical growth.
    if n > 1: # log(1) is 0, which can cause issues or flatten the start
        time_values.append(n * math.log2(n))
    else:
        time_values.append(0)

plt.figure()
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Time (n log n)")
plt.title("O(n log n) Log-Linear Time Complexity (Sorting Simulation)")
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
import math
import matplotlib.pyplot as plt

def simulate_n_log_n_operation(n):
    # This function simulates an operation with O(n log n) complexity
    # It's a conceptual representation, not a specific algorithm.
    total_operations = 0
    if n > 1:
        for _ in range(n):
            total_operations += math.log2(n)
    return total_operations

n_values = []
time_values = []

# Using a range of n values to demonstrate the growth
for i in range(1, 15):
    n = 2**i * 50 # Example input sizes (e.g., 100, 200, 400, ...)
    n_values.append(n)

    # Calculate the theoretical O(n log n) time
    if n > 1:
        time_values.append(n * math.log2(n))
    else:
        time_values.append(0)

plt.figure()
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Time (n log n)")
plt.title("O(n log n) Log-Linear Time Complexity (Conceptual Simulation)")
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
import math
import matplotlib.pyplot as plt

def conceptual_n_log_n_function(n):
    # This function simulates an operation with O(n log n) complexity
    # It performs 'n' iterations, and in each iteration, a conceptual 'log n' operation occurs.
    total_conceptual_operations = 0
    if n > 1:
        # Simulate 'n' operations, each taking log(n) time
        total_conceptual_operations = n * math.log2(n)
    return total_conceptual_operations

n_values = []
time_values = []

# Using a range of n values to demonstrate the growth
for i in range(1, 15):
    n = 2**i * 100 # Example input sizes (e.g., 200, 400, 800, ...)
    n_values.append(n)

    # Calculate the theoretical O(n log n) time
    time_values.append(conceptual_n_log_n_function(n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o')
plt.xlabel("Input size n")
plt.ylabel("Time (n log n)")
plt.title("O(n log n) Log-Linear Time Complexity (Conceptual Function)")
plt.grid(True)
plt.show()
```

```python
import time
import math
import matplotlib.pyplot as plt
import numpy as np

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Generate input sizes
n_values = np.array([10, 50, 100, 200, 500, 1000, 2000, 5000, 10000])
actual_times = []

# Measure actual execution times for quick sort
for n in n_values:
    # Create a random list of size n
    data = list(np.random.randint(0, n * 10, n))
    start_time = time.time()
    quick_sort(data)
    end_time = time.time()
    actual_times.append(end_time - start_time)

# Calculate theoretical O(n log n) times for comparison
# We'll scale it to fit the actual times for better visualization
theoretical_nlogn_times = n_values * np.log2(n_values)

# Find a scaling factor to match the theoretical curve to the actual data
# This makes the visual comparison more effective
if len(n_values) > 1 and theoretical_nlogn_times[1] > 0:
    scaling_factor = actual_times[1] / theoretical_nlogn_times[1]
else:
    scaling_factor = 1.0 # Default if scaling cannot be determined

theoretical_nlogn_times_scaled = theoretical_nlogn_times * scaling_factor

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, actual_times, label='Actual Quick Sort Time (O(n log n))', marker='o', linestyle='-')
plt.plot(n_values, theoretical_nlogn_times_scaled, label='Theoretical O(n log n) (Scaled)', linestyle='--', color='red')
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Quick Sort Time Complexity (O(n log n))")
plt.legend()
plt.grid(True)
plt.xscale('log') # Use log scale for n to better visualize complexity for larger inputs
plt.yscale('log') # Use log scale for time to better visualize complexity for larger inputs
plt.show()
```
